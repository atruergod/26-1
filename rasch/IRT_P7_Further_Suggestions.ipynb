## 4. 구조 방정식 모형(SEM)으로의 확장

### 4.1 IRT와 SEM의 연결

IRT는 SEM의 특수한 경우로 볼 수 있습니다:

```
IRT:  X_i = f(theta, beta_i) + error   [비선형 요인 모형]
SEM:  X = Lambda * eta + error          [선형 요인 모형]
```

### 4.2 EIRT를 SEM으로 확장하는 경로

```
Level 1 (Item level):
  log(P_k/P_{k-1}) = theta_n - (X_i @ gamma + epsilon_i + tau_ik)

Level 2 (Person level):
  theta_n = Z_n @ alpha + zeta_n      (= SEM person model)

Level 3 (Group level):  [SEM 확장]
  alpha_q = W_q @ delta + eta_q       (= 맥락 변수의 학교/기관 수준 효과)
```

### 4.3 Stan에서 SEM+IRT 구현 가능 여부

Stan은 완전한 Bayesian SEM+IRT 통합 모델을 구현할 수 있습니다.
구현 난이도가 높지만, 다음 패키지가 도움이 됩니다:
- **`blavaan`** (R): Bayesian SEM
- **Stan + `lavaan`**: 커스텀 SEM 구현
- **`brms`**: Multilevel SEM

### 4.4 SEM 주요 추가 개념 (IRT 이후 학습 권장)

| 개념 | 설명 | IRT와의 관계 |
|---|---|---|
| **CFA** (확인적 요인 분석) | 잠재 변수 구조 검증 | MIRT와 유사 |
| **SEM Path Analysis** | 잠재 변수 간 인과 관계 | EIRT Level 2 확장 |
| **Multilevel SEM** | 개인-집단 이중 구조 | EIRT + 학교 수준 효과 |
| **Longitudinal SEM** | 성장 곡선 모형 | 종단 IRT 확장 |
| **Bayesian SEM** | 사전 분포 통합 SEM | Stan 기반 EIRT와 동일 철학 |


````python
# 연구 주제 시각화: 복잡도 vs. 관련성 매트릭스
import matplotlib.pyplot as plt
import numpy as np

topics = [
    'MFPCM\n쓰기 채점',
    'DIF\n국적별',
    'LLTM\n문항 속성',
    '종단\n말하기',
    'CAT\n개발',
    '어휘\n난이도',
    'SEM+IRT\n통합'
]
complexity = [3, 3, 4, 4, 5, 3, 5]   # 1=easy, 5=hard
relevance  = [5, 5, 4, 4, 4, 5, 4]   # 1=low, 5=high

fig, ax = plt.subplots(figsize=(10, 7))
scatter = ax.scatter(complexity, relevance, s=300, c=range(len(topics)),
                     cmap='RdYlGn', alpha=0.8, edgecolors='black', linewidth=1.5)
for i, topic in enumerate(topics):
    ax.annotate(topic, (complexity[i], relevance[i]),
                textcoords='offset points', xytext=(10, 5), fontsize=9)

ax.axvline(3.5, color='gray', linestyle='--', alpha=0.5, label='Complexity midpoint')
ax.axhline(3.5, color='gray', linestyle=':', alpha=0.5, label='Relevance midpoint')
ax.set_xlim(1.5, 5.5); ax.set_ylim(2.5, 5.5)
ax.set_xlabel('Research Complexity (1=easy, 5=hard)', fontsize=12)
ax.set_ylabel('Relevance to Korean Language Education (1=low, 5=high)', fontsize=12)
ax.set_title('Research Topic Map: Complexity vs. Relevance\n(Upper-left: High value, manageable for Master thesis)',
             fontsize=12)
ax.text(1.8, 5.4, 'High Relevance\nLow Complexity\n-> Best for Master', ha='center',
        fontsize=9, color='green', fontweight='bold',
        bbox=dict(boxstyle='round', facecolor='lightgreen', alpha=0.3))
ax.text(4.8, 2.8, 'High Complexity\nLower Relevance\n-> PhD or later', ha='center',
        fontsize=9, color='red',
        bbox=dict(boxstyle='round', facecolor='lightyellow', alpha=0.3))
plt.tight_layout()
plt.show()
````
