# Week 5: CTT 및 Guttman 구조 분석

**학습목표 및 주요내용**
고전검사이론(CTT)의 핵심 개념 학습. Guttman 구조를 활용한 응답 패턴의 정성적 시뮬레이션 및 분석 실습.

```python
import numpy as np
import pandas as pd

# CTT Basic Statistics
def calculate_ctt_stats(responses):
    # responses: DataFrame (rows=students, cols=items)
    item_difficulty = responses.mean()
    total_scores = responses.sum(axis=1)
    item_discrimination = []
    for col in responses.columns:
        corr = responses[col].corr(total_scores)
        item_discrimination.append(corr)
    
    return item_difficulty, item_discrimination

# Guttman Pattern Simulation Placeholder
print("CTT & Guttman Analysis Section")
```

# Week 6: t-Test: 빈도주의 vs 베이지안

**학습목표 및 주요내용**
기존 논문 해석을 위한 빈도주의 t-Test(p-value) 이해. 이를 베이지안 t-Test와 비교하여 시뮬레이션 및 결과 해석 실습.

```python
from scipy import stats
import numpy as np

# Frequentist t-Test
group_a = np.random.normal(100, 15, 30)
group_b = np.random.normal(110, 15, 30)

t_stat, p_val = stats.ttest_ind(group_a, group_b)
print(f"Frequentist t-Test: t={t_stat:.3f}, p={p_val:.3f}")

# Bayesian t-Test concept would involve Stan or PyMC
print("Bayesian t-Test comparison to be implemented using Stan.")
```

# Week 7: ANOVA: 빈도주의 vs 베이지안

**학습목표 및 주요내용**
논문 상의 ANOVA 보고 방식 이해 및 베이지안 관점의 ANOVA 해석 방법 학습. 데이터 불확실성 추정 실습.

```python
# Frequentist ANOVA example
import pandas as pd
from scipy import stats

# Dummy data
data = {
    'Score': np.concatenate([np.random.normal(10, 2, 20) for _ in range(3)]),
    'Group': ['A']*20 + ['B']*20 + ['C']*20
}
df = pd.DataFrame(data)

# Simple ANOVA
groups = [df[df['Group'] == g]['Score'].values for g in ['A', 'B', 'C']]
f_stat, p_val = stats.f_oneway(*groups)
print(f"Frequentist ANOVA: F={f_stat:.3f}, p={p_val:.3f}")
```

# Week 10: IRT 모델 및 시뮬레이션

**학습목표 및 주요내용**
문항반응이론(IRT)의 수학적 배경 이해. Stan을 활용한 1PL/2PL 모델 파라미터 추정 및 시각화 실습.

```python
# IRT 1PL/2PL Stan Model Structure
irt_code = """
data {
  int<lower=0> J; // students
  int<lower=0> K; // items
  int<lower=0> N; // observations
  array[N] int<lower=1, upper=J> jj;
  array[N] int<lower=1, upper=K> kk;
  array[N] int<lower=0, upper=1> y;
}
parameters {
  vector[J] theta;      // student ability
  vector[K] beta;       // item difficulty
  vector<lower=0>[K] alpha; // item discrimination (for 2PL)
}
model {
  theta ~ normal(0, 1);
  beta ~ normal(0, 1);
  alpha ~ lognormal(0, 0.5);
  y ~ bernoulli_logit(alpha[kk] .* (theta[jj] - beta[kk]));
}
"""

print("IRT 2PL Model Stan code defined.")
```
